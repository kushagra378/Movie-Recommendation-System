# TRAINING MODEL

```python
netflix_data = pd.read_csv('netflix.csv')
hotstar_data = pd.read_csv('hotstar.csv')
prime_data = pd.read_csv('prime.csv')

df1_netflix = pd.DataFrame(netflix_data)

df1_hotstar = pd.DataFrame(hotstar_data)

df1_prime = pd.DataFrame(prime_data)
```

# Count Vectorizer

## NETFLIX

```python
cv = CountVectorizer()
count_matrix_netflix = cv.fit_transform(df1_netflix['listed_in'])
count_array = count_matrix_netflix.toarray()
df2_netflix = pd.DataFrame(data=count_array,columns = cv.get_feature_names())

tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True) 
tfidf_transformer.fit(df2_netflix)

# print idf values 
df_idf = pd.DataFrame(tfidf_transformer.idf_, index=cv.get_feature_names(),columns=["idf_weights"]) 
# sort ascending 
df_idf.sort_values(by=['idf_weights'])

# count matrix 
count_vector=cv.transform(df_idf) 
# tf-idf scores 
tf_idf_vector=tfidf_transformer.transform(count_vector)

feature_names = cv.get_feature_names() 
#get tfidf vector for first document 
first_document_vector=tf_idf_vector[0] 
#print the scores 
df3_netflix = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"]) 
df3_netflix.sort_values(by=["tfidf"],ascending=False)

# Initialize an instance of tf-idf Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Generate the tf-idf vectors for the corpus
tfidf_matrix = tfidf_vectorizer.fit_transform(df1_netflix['listed_in'])

# compute and print the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

indices = pd.Series(df1_netflix.index, index=df1_netflix['title']).drop_duplicates()
movie_plots_netflix = df1_netflix['listed_in']
```

## HOTSTAR

```python
count_matrix_hotstar = cv.fit_transform(df1_hotstar['listed_in'])
count_array_hotstar = count_matrix_hotstar.toarray()
df2_hotstar = pd.DataFrame(data=count_array_hotstar,columns = cv.get_feature_names())

tfidf_transformer_hotstar=TfidfTransformer(smooth_idf=True,use_idf=True) 
tfidf_transformer_hotstar.fit(df2_hotstar)

# print idf values 
df_idf_hotstar = pd.DataFrame(tfidf_transformer_hotstar.idf_, index=cv.get_feature_names(),columns=["idf_weights"]) 
# sort ascending 
df_idf_hotstar.sort_values(by=['idf_weights'])

# count matrix 
count_vector_hotstar=cv.transform(df_idf_hotstar) 
# tf-idf scores 
tf_idf_vector_hotstar=tfidf_transformer_hotstar.transform(count_vector_hotstar)

feature_names_hotstar = cv.get_feature_names() 
#get tfidf vector for first document 
first_document_vector_hotstar=tf_idf_vector_hotstar[0] 
#print the scores 
df3_hotstar = pd.DataFrame(first_document_vector_hotstar.T.todense(), index=feature_names_hotstar, columns=["tfidf"]) 
df3_hotstar.sort_values(by=["tfidf"],ascending=False)

# Initialize an instance of tf-idf Vectorizer
tfidf_vectorizer_hotstar = TfidfVectorizer()

# Generate the tf-idf vectors for the corpus
tfidf_matrix_hotstar = tfidf_vectorizer_hotstar.fit_transform(df1_hotstar['listed_in'])

# compute and print the cosine similarity matrix
cosine_sim_hotstar = cosine_similarity(tfidf_matrix_hotstar, tfidf_matrix_hotstar)

indices_hotstar = pd.Series(df1_hotstar.index, index=df1_hotstar['title']).drop_duplicates()
movie_plots_hotstar = df1_hotstar['listed_in']
```

```python
imdb_rating=[]
img_poster=[]
def outputimage(k):
    for i in range(0,len(k)):
        URL = k[i]
        r = requests.get(url=URL)
        soup = BeautifulSoup(r.text, 'html.parser')
        image = soup.find("div", class_="ipc-media ipc-media--poster-27x40 ipc-image-media-ratio--poster-27x40 ipc-media--baseAlt ipc-media--poster-l ipc-poster__poster-image ipc-media__img").img['src']
        urllib.request.urlretrieve(image,"poster"+str(i)+".png")
        img_poster.append(Image.open("poster"+str(i)+".png"))
        rating = soup.find("span",class_="sc-7ab21ed2-1 jGRxWM").text
        imdb_rating.append(rating)
        print("Downloading Poster "+str(i+1)+"...")
```
